## Duration Formatting

```python
#| export
def format_duration_mmss(
    seconds: Optional[float]  # Duration in seconds
) -> str:  # Formatted string (mm:ss)
    """Format duration in seconds as mm:ss for summary display."""
    if seconds is None:
        return "--:--"
    minutes = int(seconds // 60)
    secs = int(seconds % 60)
    return f"{minutes}:{secs:02d}"
```

```python
assert format_duration_mmss(0.0) == "0:00"
assert format_duration_mmss(65.5) == "1:05"
assert format_duration_mmss(2537.0) == "42:17"
assert format_duration_mmss(None) == "--:--"
print("format_duration_mmss tests passed")
```

```python
#| export
def format_duration_seconds(
    seconds: Optional[float]  # Duration in seconds
) -> str:  # Formatted string (e.g., "10.3s")
    """Format duration in seconds as Xs or X.Xs for average display."""
    if seconds is None:
        return "-.-s"
    return f"{seconds:.1f}s"
```

```python
assert format_duration_seconds(10.3) == "10.3s"
assert format_duration_seconds(0.0) == "0.0s"
assert format_duration_seconds(65.25) == "65.2s"
assert format_duration_seconds(None) == "-.-s"
print("format_duration_seconds tests passed")
```

```python
#| export
def format_time_range(
    start: Optional[float],  # Start time in seconds
    end: Optional[float]  # End time in seconds
) -> str:  # Formatted range (e.g., "0.0s - 2.1s")
    """Format time range for sample segment display."""
    if start is None or end is None:
        return "-.-s - -.-s"
    return f"{start:.1f}s - {end:.1f}s"
```

```python
assert format_time_range(0.0, 2.1) == "0.0s - 2.1s"
assert format_time_range(2512.0, 2537.0) == "2512.0s - 2537.0s"
assert format_time_range(None, 2.1) == "-.-s - -.-s"
print("format_time_range tests passed")
```

## Text Formatting

```python
#| export
def truncate_text(
    text: Optional[str],  # Full text to truncate
    max_length: int = 60  # Maximum length before truncation
) -> str:  # Truncated text with ellipsis if needed
    """Truncate text for sample segment display."""
    if text is None:
        return ""
    if len(text) <= max_length:
        return text
    return f"{text[:max_length-3]}..."
```

```python
assert truncate_text("Short text") == "Short text"
assert truncate_text("A" * 60) == "A" * 60
assert truncate_text("A" * 70) == "A" * 57 + "..."
assert truncate_text(None) == ""
assert truncate_text("Hello world", max_length=8) == "Hello..."
print("truncate_text tests passed")
```
